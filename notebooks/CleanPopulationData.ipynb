```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
pop=pd.read_excel("../data/population_by_area.xls",sheet_name="Persons")
```

```python
pop=pop.drop(index=np.arange(0,5))
```

```python
pop.columns=pop.iloc[0].to_list()
```

```python
pop.index=np.arange(0,len(pop))
```

```python
pop=pop.drop(index=0)
```

```python
popAll=pop[pop['AGE GROUP'].str.contains("All")]
```

```python
popAA=popAll[~popAll['AREA'].str.match('England')]
```

```python
popAA=popAA.drop(columns=popAA.columns[6:None])
popAA=popAA.drop(columns='AGE GROUP')
```

```python
popAA.columns=["Area code","Area name","pop2018","pop2019","pop2020"]
```

```python
popAA.index=np.arange(0,len(popAA))
```
